```python
import numpy as np
n = 1000000
suma_float = np.float32(0.0)
suma_double = np.float64(0.0)
for i in range(n):
    suma_float += np.float32(0.1)
    suma_double += np.float64(0.1)
valor_teorico = n * 0.1
# Errores
error_float = abs(valor_teorico - suma_float)
error_double = abs(valor_teorico - suma_double)
print("Valor teórico:", valor_teorico)
print("\n======== float32 ========")
print("Resultado:", suma_float)
print("Error:", error_float)
print("\n======== float64 ========")
print("Resultado:", suma_double)
print("Error:", error_double)
```
