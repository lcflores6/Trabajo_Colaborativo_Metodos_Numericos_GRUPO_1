```python
# Valor real y aproximado
import math
pi_real = math.pi
pi_aprox = 3.1416
# Cálculo de errores
error_abs = abs(pi_real - pi_aprox)
error_rel = error_abs / abs(pi_real)
print("Valor real:", pi_real)
print("Valor aproximado:", pi_aprox)
print("Error absoluto:", error_abs)
print("Error relativo:", error_rel)
```
